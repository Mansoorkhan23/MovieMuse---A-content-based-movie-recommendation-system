```python
import ast
```

Creating a helper function:

**Converting string of List to List**: 

Using function literal eval in Module Ast to do this task:

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies['genres'] = movies['genres'].apply(convert)

movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)

movies.head()
```

```python
movies['cast'][0]
```

- Taking the **top 3 actors** from all the movies

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)

movies.head()
```

```python
# movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

- We will take the only values, where **Job =  Director**;

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

- The problem if we allow spaces in our data is that for ex: if we have one cast name - "Sam Worthington" and one crew name - "Sam Mendes" and when the tags are created 'Sam' becomes one entity and it points out to 2 people so recommendations would go wrong; So, **We remove spaces**

- Finally, 'SamWorthington' will be a single entity and 'SamMendes' will be a single entity

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

Creating a new column '**tags**' - which is a combination of the all the 4 columns 'genres', 'overview', 'keywords', 'cast', 'crew'

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(5)
```

Dropping the other unnecessay columns -

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

**Converting 'tags' to string format**

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

lower case conversion -

```python
new['tags'] = new['tags'].apply(lambda x:x.lower())
new.head()
```

- **Performing Text Vectorization** and **removing english stopwords** from our data (are, is, and, etc)

- Text to vector conversion: called Text Vectorization and technique used is: Bag Of Words

- In Bag Of Words: we combine all the tags (tag1 + tag2 +....and so on) 

- In this big string obtained after combining, we take the 5000 most common words whose frequency is more, example:

In Movie 1:

- Action |  Future |      Love | .... | Hate |
- Word1 | Word2 | Word3 | .... | Word5000 |

-    5 |   3   |   0 | ...    |      1 |

.

In Movie 2:
-    2 |   0   |  1 | ...    |      3 |

.


Similarly in Movie 5000:
-    8 |   3   |    4 | ...    |      0 |

.

Finally, we get a table of shape (5000,5000) that is every movie is a vector and we will have 5000 vectors (5000 dimensional space)

- we will remove stopwords like is, are, and, etc as these won't contribute any meaning to the recommendations using sckit learn library (countvectorizer)


```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

- Transforming 'tags' column to array

```python
vector = cv.fit_transform(new['tags']).toarray()
```

- After transformation:

```python
cv.get_feature_names()
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))
  
  return " ".join(y)
```

```python
new['tags'] = new['tags'].apply(stem)
```

- Our final task is to **calculate the distance** from one movie to another movie; The more the distance, the lesser the similarity

- For that we will be using **cosine similarity** to calculate the distance between the movies (which is calc. the angle between the vectors) unlike euclidean distance where we are calculating the distance between the tips 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

- Performing Cosine similarity:

```python
similarity = cosine_similarity(vector)
```

- Calculating the distance of each movie with 4806 movies -

```python
 cosine_similarity(vector).shape
```

```python
similarity
```

```python
similarity[0]
```

- Next, we have to **sort the movies in descending order** w.r.t the distance in order to have the movies with **highest cosine similarity at the top**: but in this process we loose the index value of the movie to avoid we will use the function 'enumerate':

- Ex: Distance of 0th movie with all the movies:

```python
list(enumerate(similarity[0]))
```

- Next, we will do sorting w.r.t the 2nd column i.e., the similarity distance in descending order

```python
sorted(list(enumerate(similarity[0])),reverse=True,key = lambda x: x[1])
```

- **Fetching the index position of the movie to be searched:**

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

- And finally, we will **fetch the title of top 5 movies**

```python
def recommend(movie):  
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

- **Final Top 5 Recommended movies** for the selected movie -

```python
recommend('Gandhi')
```
